```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def convert_image(in_image, out_image):

  # Open the image
  image = Image.open(in_image)
  # Convert the image to a desired bit depth
  # "L" (8-bit grayscale), "RGB" (24-bit color), or "RGBA" (32-bit color with alpha channel)
  new_image = image.convert("RGB")
  # Resize the image to a specific width and height
  new_size = (428, 428)  # New width and height
  resized_image = new_image.resize(new_size)

  # Save the modified image
  resized_image.save(out_image)

def plot_image(in_image):
  """
    :param image: the image to be plotted in a 3-D matrix format
    image = np.fromfile(image_file, dtype=np.uint8, count=SIZE)
    # force into image matrix
    image = np.reshape(image, (3, 96, 96))

    :return: None
  """
  image = Image.open(in_image)
  plt.imshow(image)
  plt.show()
```
